```python
# read the csv file using pandas
df = pd.read_csv("reviews.csv")
df.head()
```

```python
# checking for missing values
df.info()
```

```python
#df['score'].unique()
df.score.unique()
```

```python
import matplotlib.pyplot as plt
plt.hist(df.score)
```

```python
def convert_score_to_sentiment(rating):
  if rating <= 2:
    return 0
  elif rating == 3:
    return 1
  else:
    return 2

df['sentiment'] = df.score.apply(convert_score_to_sentiment)
```

```python
df.head()
```

```python
# distribution of data in sentiment column
plt.hist(df.sentiment)
```

```python
## CREATE DATASET
from sklearn.model_selection import train_test_split

# trainset
df_train, df_test = train_test_split(df, test_size=0.2, random_state=42)
```

```python
# import tfidfvectorization library to convert dataset into vectors
from sklearn.feature_extraction.text import TfidfVectorizer


tfidfvector = TfidfVectorizer(ngram_range=(2, 2))
vectorized_trainset = tfidfvector.fit_transform(df_train.content)
```

```python
# import Naive Bayes on the content and sentiment
from sklearn.naive_bayes import MultinomialNB

naive = MultinomialNB()
naive.fit(vectorized_trainset, df_train.sentiment)
```

```python
## TESTING THE MODEL
#tfidfvector = TfidfVectorizer(ngram_range=(2, 2))
vectorized_testset = tfidfvector.transform(df_test.content)
predictions = naive.predict(vectorized_testset)
```

```python
## EVALUATION
from sklearn.metrics import confusion_matrix, accuracy_score


matrix = confusion_matrix(df_test.sentiment, predictions)
print(matrix)
accuracy = accuracy_score(df_test.sentiment, predictions)
print(accuracy)
```
